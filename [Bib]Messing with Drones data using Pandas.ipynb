# Warehouse Inventories

Here we put the warehouse stock into one list of all the stock values

```python
num_whs = 10
wh_stock_list = []

for wh in range(num_whs):
    wh_stock_list.append(data[5 + 2*wh])
wh_stock_list
```

Split up the warehouse stock (wh_stock) into a dictionary

```python
wh_stock_dict = { i: wh_stock_list[i] for i in range(len(wh_stock_list)) }
wh_stock_dict[0]
```

Change the stock values from a string to a list

```python
for i in range(10):
    wh_stock_dict[i] = wh_stock_dict[i].split()
wh_stock_dict
```

Convert the stock values from a dictionary to a data frame

```python
df = pd.DataFrame(wh_stock_dict)
df
```

Convert values in the tables to integers

```python
df = df.apply(pd.to_numeric)
```

Add a row that sums the total stock count of each warehouse

```python
df.loc['Warehouse total stock'] =  df.sum()
df
```

Add a column that sums the total stock count of each item across all warehouses

```python
df['Total item stock'] =  df.sum(axis=1)
df
```
